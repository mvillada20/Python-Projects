## Rename the three main columns I will be comparing data for: LocationDesc will be renamed to "Locations," Topic will be renamed to "Diseases," and StratificationID1 will be renamed to "Ethnicity"

```python
df.rename(columns={'LocationDesc': 'Locations', 'Topic': 'Diseases', 'StratificationID1': 'Ethnicity'}, inplace=True)
df
```

## Continuing the clean up of the data: The variable 'HIS' under the "Ethnicity" column will be renamed to 'HISPANIC' (full word)

```python
df['Ethnicity'] = df['Ethnicity'].replace(['HIS'],['HISPANIC'])
df
```

## Continuing the clean up of the data: Dropping various columns. We started with 34 columns and after the clean up, we are down to 18 columns

```python
df1=df.drop(['LocationAbbr', 'DataSource', 'Response', 'LocationID', 'TopicID', 'QuestionID', 'DataValueTypeID', 'StratificationCategoryID2', 'StratificationID2', 'StratificationCategoryID3', 'StratificationID3', 'StratificationCategory2', 'Stratification2', 'StratificationCategory3', 'Stratification3', 'GeoLocation'], axis=1)
df1
```

## Identify the various diseases and ethnicities under the Disease and Ethnicity column. Due to the large dataset, all the information cannot be displayed properly on the chart

```python
df1.Diseases.unique()
```

```python
df1.Ethnicity.unique()
```

## After completing the clean up of the raw data, I want to filter the data to only display information from New York (no other states or locations)

```python
df2=df1[df1['Locations'] == 'New York']
df2
```

## To further filter, I want to display the number of cases of Diabetes in New York. Per the information below, there are (2,098) cases of people with a Diabetes diagnosis in New York - (including all ethnicities)

```python
df3=df2[df2['Diseases'] == 'Diabetes']
df3
```

## To reinforce the information above, please see the chart below to identify the number of cases for each disease including Diabetes (2,098) and others.

```python
NewYork=df2.groupby(['Diseases'])['Diseases'].count().sort_values(ascending=False)
NewYork
```

## I want to continue to compare data. I want to look at the number of Hispanic people with Diabetes in New York and according to the data below, there are (264) Hispanic people in New York with a Diabetes diagnosis.

```python
df4=df3[df3['Ethnicity'] == 'HISPANIC']
df4
```

## I wanted to identify the number of cases of Diabetes, grouped by Ethnicity. Also, the information below reinforces the information above, in chart form. Below you can identify the number of people with Diabetes grouped by Ethnicity.

```python
ny=df3.groupby(['Ethnicity'])['Diseases'].count().sort_values(ascending=False)
ny
```

## Further consolidating the data. (264) Hispanics with Diabetes in New York.

```python
ny1=df4.groupby(['Ethnicity'])['Diseases'].count().sort_values(ascending=False)
ny
```

## Below is the above information in pie chart form

```python
ny.plot.pie()
```
