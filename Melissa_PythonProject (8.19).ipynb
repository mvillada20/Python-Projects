## Import packages and dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

## Pulling the raw data. There are 212331 rows and 15 columns of information

```python
df
```

```python
df.describe()
```

## Forcing non-numerical values to become NaNs on the Mean Score column. Changing NA and 'S' values to NaN.

```python
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors = 'coerce')
df
```

## Dropping any non-numerical value on the Mean Score column. It skewed the data, but I was able to work with only numbers, no N/A or NaN values.

```python
df.dropna(inplace=True)
```

```python
df
```

## On line #4 you will see the descriptive statistics for the raw data, however, line #8 will display the data after the clean-up (removing NaN's from the Mean Score column). The information is a bit skewed, but I found it easier to work without them and work with only numbers.

```python
df.describe()
```

## I chose to research information on Francis Lewis High School, so I filtered particularily by that school name. There are 449 rows and 15 columns of information after removing the NaNs. Without removing the non-numerical values there would be 490 rows.

```python
df1=df[df['School Name'] == 'Francis Lewis High School']
df1
```

## I went a little further and filtered by a particular Regents Exam - U.S. History and Government.

```python
df2=df1[df1['Regents Exam'] == 'U.S. History and Government']
df2
```

## Additional descriptive statistics, but for FLHS only! I decided to use the information available for Francis Lewis High School and group by Regents Exam and Year and filtered to the max Mean Score. That is the information desplayed below; max amount of Mean Score for that particular year and regents topic. As you can see, there are all numerical values - without dropping the NaNs, there would be a bunch of 'S' shown.

```python
df1.groupby(['Regents Exam','Year'])['Mean Score'].max()
```

## Below, I wanted to dive a little deeper and groupby the particular regents exam (U.S. History and Government), the year and the max Mean Score for that year.

```python
ms=df2.groupby(['Regents Exam','Year'])['Mean Score'].max()
ms[0]
```

```python
ms
```

## Below is a bar graph of the above information - it is showing the max Mean Scores for the U.S. History and Government regents exams between 2015-2017 for FLHS only.

```python
ms=ms.astype(float)
ms.plot.bar(color='red')
```

```python
df['Mean Score']
```

## Below - Descriptive statistical data demonstrating the cumulative average Mean Score for the entire school's (FLHS) various Regents Exams, and sorted from highest to lowest.

```python
flhs=df1.groupby(['School Name','Regents Exam'])['Mean Score'].mean().sort_values(ascending=False)
flhs
```

## The below graph visually represents line 28 by using a bar graph that shows the cumulative average Mean Score for all the regents exam in FLHS.

```python
flhs.plot.bar(color={'red','green','orange','blue'})
```

## Below demonstrates the overall information of ALL the schools grouped by regents exam, with the max Mean Score

```python
df.groupby(['School Name','Regents Exam'])['Mean Score'].mean().sort_values(ascending=False)
```
